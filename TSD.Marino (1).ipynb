```python
#Import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

#csv file 
earthquakes = pd.read_csv("earthquake_data.csv", parse_dates=['US Region'])
```

## Who doesn't love a good History channel documentary?
When I was a kid, I watched documentaries all the time. One of the most common topics on the History channel was the possibility of a catastrophic disaster and what the consequences would look like in today's world. The data collected in this set reminds me of these documentaries, but brings in a more realistic take on how the population feels about the idea of a minor or massive earthquake taking place. 
This data set takes a look at the population of the US and asks questions pertaining to the threat of the San Andreas fault annd earthquakes in general in everyday life. There are a total of 1,014 responses.

## 1. How does familiarity with the San Andreas fault line affect nervousness about the possibility of a big, catastrophic earthquake?

My first question focues on familiarity and nervousness of the general population in terms of the possibility of a large earthquake or natural disaster. The results are shown below.

```python
#Familiarity and Nervousness
question1 = pd.crosstab(earthquakes['How familiar are you with the San Andreas Fault line?'], earthquakes['How worried are you about the Big One, a massive, catastrophic earthquake?'])

#Visualization
question1 = question1.cumsum()
plt.clf()
question1.plot(kind = 'barh', figsize = (10,10), colormap = 'autumn')
plt.title('Familiarity and Nervousness')
plt.xlabel('Frequency of Answer')
```

As seen above, the more familiar one is with the fault line and its capabilities, the more nervous they are about the possibility of it actually occuring, along with the effects that would come with it. Intriguingly, most people (no matter how familiar with the fault line) are categorized in "Not at all worried" or "Not so worried" about a large event occuring in the near future. Few people are "Extremely familiar" with the San Andreas fault, but those are resignate with the "Extremely worried" category in a higher percentage than those who are any of the other familiarities.  

## 2. How does U.S. Region affect nervousness about the occurance of earthquakes in general?
For this question, I looked at the participants proximity to the San Andreas fault line and how it related to nervousness about a massive earthquake occurring.

```python
#Region and Nervousness Broad
question2 = pd.crosstab(earthquakes['US Region'], earthquakes['How worried are you about the Big One, a massive, catastrophic earthquake?'])

#Visualization
question2 = question2.cumsum()
plt.clf()
question2.plot(kind = 'barh', figsize = (10,15), colormap = 'cool')
plt.title('Nervousness Depending on US Region')
plt.xlabel('Frequency of Answer')
```

The results above reveal that the closer a participant lives to the fault line, the more worried he/she is worried about the San Andreas going off. The west coast had the highest levels of nervousness and non-coincidentally is closest to the fault line. 

## 3. Do age and potentiality for experiencing one or more earthquakes correlate?

My final question looks at age and its relation to earthquake experience. The graph below shows a correlation.

```python
#Age and Earthquake Experience
question3 = pd.crosstab(earthquakes['Age'], earthquakes['Have you ever experienced an earthquake?'])

#Visualization
question3 = question3.cumsum()
plt.clf()
question3.plot(kind = 'bar', figsize = (10,8), colormap = 'winter')
plt.title('Age and Earthquake Experience')
plt.ylabel('Frequency of Answer')
```

The results above display the correlation between aging and the possibility of experiencing an earthquake. Older participants have a greater chance of having experienced an earthquake (minor or major) in their lifetime. However, minor earthquakes were much more common. 

## Problems with Data
Although the data asked good, detailed questions of the participants, the survey only had 1,014 answers that were spread out across the country. This is not a good, varied number for amount of people who are affected by the idea of earthquakes. Therefore, the values given may not be reflective of the entire population.

The subject of a catastrophic event is popular in the US culture, as it seems like an impossibility. That's what makes it interesting.
